```python

# 拆分測試組20% 及驗證組80%
import pandas as pd
from sklearn.model_selection import train_test_split

# 讀取數據
df = pd.read_csv("/content/drive/MyDrive/電信客戶流失/Fixed-Telco-Dtype.csv")

# **分割 80% 訓練集 / 20% 測試集**
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Churn"]), df["Churn"],test_size=0.2, random_state=42, stratify=df["Churn"])

# **將訓練 & 測試數據合併**
train_data = pd.concat([X_train, y_train], axis=1)
test_data = pd.concat([X_test, y_test], axis=1)

# **存檔**
train_data.to_csv("/content/drive/MyDrive/電信客戶流失/Telco_train.csv", index=False)  # 訓練集
test_data.to_csv("/content/drive/MyDrive/電信客戶流失/Telco_test.csv", index=False)   # 測試集

print("訓練 & 測試數據已分割並存檔！")
```
